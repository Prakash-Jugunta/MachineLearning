```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV ,RandomizedSearchCV
import logging
logging.basicConfig(level=logging.DEBUG) 
```

# A1

```python
data=pd.read_csv('breastCancer.csv')
data.info()
```

```python
# Class changer for apply function
def classChanger(val):
    if val=='M':
        return 0
    return 1
X=data.drop(columns=['diagnosis','id'])
X=pd.DataFrame(StandardScaler().fit_transform(data.drop(columns=['diagnosis','id'])),columns=X.columns)
Y=data['diagnosis'].apply(classChanger)
X
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
neigh=KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train,Y_train)
y_train_pred = neigh.predict(X_train)
y_test_pred = neigh.predict(X_test)
```

```python
logging.info("Train Confusion Matrix:")
logging.info(confusion_matrix(Y_train, y_train_pred))

logging.info("\nTest Confusion Matrix:")
logging.info(confusion_matrix(Y_test, y_test_pred))

logging.info("\nTrain Classification Report:")
logging.info(classification_report(Y_train, y_train_pred))

logging.info("\nTest Classification Report:")
logging.info(classification_report(Y_test, y_test_pred))

train_report = classification_report(Y_train, y_train_pred, output_dict=True)
test_report = classification_report(Y_test, y_test_pred, output_dict=True)

train_f1 = train_report["weighted avg"]["f1-score"]
test_f1 = test_report["weighted avg"]["f1-score"]

if train_f1 > test_f1 + 0.1:  
    logging.info("\nInference: The model is overfitting (high performance on train data but low on test).")
elif train_f1 < 0.6 and test_f1 < 0.6:  
    logging.info("\nInference: The model is underfitting (low performance on both train and test data).")
else:
    logging.info("\nInference: The model is well-fitted (similar performance on both train and test data).")
```

# A2

```python
data=pd.read_csv("purchaseData.csv")
data
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, r2_score

A = data.drop(columns=['Customer', 'Payment (Rs)']) 
B = data['Payment (Rs)'] 
pseudo_inverse = np.linalg.pinv(A)
X = pseudo_inverse @ B

B_pred = A @ X  
mse = mean_squared_error(B, B_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(B, B_pred)
r2 = r2_score(B, B_pred)

logging.info(f"Mean Squared Error (MSE): {mse:.4f}")
logging.info(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
logging.info(f"Mean Absolute Percentage Error (MAPE): {mape * 100:.2f}%")
logging.info(f"R² Score: {r2:.4f}")
if r2 > 0.9:
    logging.info("\nInference: The model has a strong predictive performance.")
elif r2 > 0.7:
    logging.info("\nInference: The model performs well but can be improved.")
elif r2 > 0.5:
    logging.info("\nInference: The model has moderate performance and may need better feature selection.")
else:
    logging.info("\nInference: The model has weak predictive power, indicating potential underfitting.")
```

```python
#uniform distruction
X = np.random.uniform(1,10,20)
Y = np.random.uniform(1,10,20)

classes = np.array([1 if x + y > 10 else 0 for x, y in zip(X, Y)]) 
colors = np.array(['blue', 'red'])[classes]  
def scatter_plot(X,Y,colors):
  plt.scatter(X,Y,c=colors)
  plt.xlabel("feature X")
  plt.ylabel("feature Y")
  plt.show()

scatter_plot(X,Y,colors)
```

# A4

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Assume X, Y, and classes are already defined
train_data = np.column_stack((X, Y))

# kNN Classifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(train_data, classes)

# Generate test set
X_test, Y_test = np.meshgrid(np.arange(0, 10.1, 0.1), np.arange(0, 10.1, 0.1))
test_points = np.c_[X_test.ravel(), Y_test.ravel()]

# Predict classes for test set
test_classes = neigh.predict(test_points)

# Convert class labels to colors (Assuming classes are 0 and 1)
colors = np.array(['blue', 'red'])[test_classes]

# Scatter plot
plt.scatter(test_points[:, 0], test_points[:, 1], c=colors, s=2)  # Small dot size for better visibility
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("kNN Classification Results")
plt.show()
```

# A5

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Assume X, Y, and classes are already defined
train_data = np.column_stack((X, Y))

# Generate test set
X_test, Y_test = np.meshgrid(np.arange(0, 10.1, 0.1), np.arange(0, 10.1, 0.1))
test_points = np.c_[X_test.ravel(), Y_test.ravel()]

# Different k values to test
k_values = [1, 3, 5, 10]

plt.figure(figsize=(15, 10))  # Set figure size

for i, k in enumerate(k_values, 1):
    # Train kNN with k neighbors
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(train_data, classes)
    
    # Predict test set classes
    test_classes = neigh.predict(test_points)
    
    # Convert class labels to colors (Assuming classes are 0 and 1)
    colors = np.array(['blue', 'red'])[test_classes]

    # Plot results
    plt.subplot(2, 3, i)  # Create subplots in a grid (2 rows, 3 columns)
    plt.scatter(test_points[:, 0], test_points[:, 1], c=colors, s=2)  # Small dot size
    plt.scatter(X, Y, c=np.array(['blue', 'red'])[classes], edgecolors='black', marker='o')  # Plot training points
    plt.title(f"k = {k}")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")

plt.tight_layout()
plt.show()
```

# A6

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Load data
data = pd.read_csv('breastCancer.csv')

# Feature selection
feature1 = 'radius_mean'
feature2 = 'texture_mean'

# Convert diagnosis to numerical class labels
def classChanger(val):
    return 0 if val == 'M' else 1

X = data[[feature1, feature2]].values
y = data['diagnosis'].apply(classChanger).values

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# A3: Scatter plot of training data
plt.figure(figsize=(6,5))
plt.scatter(X_train[:, 0], X_train[:, 1], c=np.array(['blue', 'red'])[y_train], edgecolors='black')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title("Training Data Scatter Plot")
plt.show()

# A4: Generate test set
X_test_grid, Y_test_grid = np.meshgrid(np.arange(X[:, 0].min(), X[:, 0].max(), 0.1),
                                       np.arange(X[:, 1].min(), X[:, 1].max(), 0.1))
test_points = np.c_[X_test_grid.ravel(), Y_test_grid.ravel()]

# Train kNN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
test_classes = knn.predict(test_points)

# Scatter plot for test data classification
plt.figure(figsize=(6,5))
plt.scatter(test_points[:, 0], test_points[:, 1], c=np.array(['blue', 'red'])[test_classes], s=2)
plt.scatter(X_train[:, 0], X_train[:, 1], c=np.array(['blue', 'red'])[y_train], edgecolors='black', marker='o')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.title("kNN Classification (k=3)")
plt.show()

# A5: Repeat for multiple k values
k_values = [1, 3, 5, 10, 15]

plt.figure(figsize=(15,10))
for i, k in enumerate(k_values, 1):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    test_classes = knn.predict(test_points)
    
    plt.subplot(2, 3, i)
    plt.scatter(test_points[:, 0], test_points[:, 1], c=np.array(['blue', 'red'])[test_classes], s=2)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=np.array(['blue', 'red'])[y_train], edgecolors='black', marker='o')
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.title(f"k = {k}")

plt.tight_layout()
plt.show()
```

# A7

```python
param_grid = {'n_neighbors': list(range(1,11))} 
neigh = KNeighborsClassifier()

grid_search = GridSearchCV(neigh, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, Y_train)

print(f"Best k (Randomized): { grid_search.best_params_['n_neighbors'] }")

random_search = RandomizedSearchCV(neigh, param_grid, n_iter=10, cv=5, scoring='accuracy', random_state=42, n_jobs=-1)
random_search.fit(X_train, Y_train)

print(f"Best k (Randomized): {random_search.best_params_['n_neighbors']}")

```
